```python
import pandas as pd
import numpy as np
import json
from pathlib import Path 
```

```python
# list dei paesi
countries = ["CA", "DE", "FR", "GB", "IN", "JP", "KR", "MX", "RU", "US"]

# cartella data contiene tutti i file json e csv ed è nella stessa dir del notebook
DATA_DIR = Path("data")

#creiamo i dizionari dove le Chiavi saranno gli acronimi dei paesi 
#e i loro contenuti saranno i dataframe corrispondenti
csv_dfs = {}
json_data = {}

#ciclo for per leggere tutti i file in automatico (invece di farli tutti a mano dato che hanno lo stesso nome)
for country in countries: #per ogni paese della lista creiamo il percorso del file
    csv_path = DATA_DIR / f"{country}videos.csv" 
    json_path = DATA_DIR / f"{country}_category_id.json"
    
    # Lettura dei CSV 
    df = pd.read_csv(csv_path, encoding="latin1", encoding_errors="replace") 
    #latin1 evita errori di lettura. i file youtube non sono utf-8
    #encoding_errors="replace" sostituisce un carattere anomalo con un altro sismbolo speciale
    #per evitare che crashi
    csv_dfs[country] = df #dizionario con chiave = country, valore = dataframe
    
    # Lettura JSON (punto 14)
    with open(json_path, encoding="utf-8") as f: #open apre il file json
        j = json.load(f) #lo trasforma in un dizionario python
    json_data[country] = j

print("Import completato")
```

```python
#ora creiamo il nuovo data frame:

# Creiamo una lista vuota che useremo per salvare i vari dataframe
dfs_wc = []

# Per ogni paese (country) e per ogni dataframe associato...
for country, df in csv_dfs.items():
    
    # 1. Facciamo una copia del dataframe originale
    #    (così evitiamo di modificare accidentalmente quello salvato nel dizionario)
    temp = df.copy()
    
    # 2. Aggiungiamo una colonna nuova chiamata "country"
    #    e come valore mettiamo il nome del paese (es: "CA", "JP", "US", ecc.)
    temp["country"] = country
    
    # 3. Mettiamo questo dataframe modificato dentro la lista
    dfs_wc.append(temp)

# 4. Uniamo tutti i dataframe della lista in un unico grande dataframe
df2 = pd.concat(dfs_wc, ignore_index=True) #ignore_index

# 5. Mostriamo qualche informazione
print("Concatenazione completata. Shape:", df2.shape)
df2.head()  #check
```

```python
df2['publish_time_dt'] = pd.to_datetime(df2['publish_time'], errors='coerce')

# errors = 'coerece' Se un valore non può essere convertito in data-ora, trasformalo in NaT (Not a Time) invece di dare errore. 
# creiamo una nuova colonna con il tempo di pubblicazione 
#convertito in formato datetime.

df2[['publish_time_dt']].head()
```

```python
# stabiliamo l'Inizio e fine dell’intervallo da 10 minuti
start = df2['publish_time_dt'].dt.floor('10T') 
##prende ogni orario di pubblicazione e lo arrotonda verso
# il basso al blocco di 10 minuti più vicino.
#10T = 10 minuti
#floor arrotonda verso il basso temporalmente al multiplo di 10 piu vicino
end = start + pd.Timedelta(minutes=10)
#aggiunge 10 minuti al valore di start per ottenere la fine dell’intervallo.
#pd.Timedelta(minutes=10) crea un oggetto “durata temporale”
#  di 10 minuti.  Sommarlo allo start crea una variabile
#  che rappresenta l’estremo superiore dell’intervallo.


#  Etichettiamo l'intervallo, solo con ore e minuti
df2['publish_time_interval'] = start.dt.strftime('%H:%M') + ' - ' + end.dt.strftime('%H:%M')
```

```python
#7. For each interval, determine the number of videos,
#  average number of likes and of dislikes.
```

```python
interval_stats = df2.groupby('publish_time_interval').agg(
    num_videos=('publish_time_dt', 'count'),
    avg_likes=('likes', 'mean'),
    avg_dislikes=('dislikes', 'mean')
).reset_index()

#prima raggruppiamo per publish_time_interval
#successivamente con .agg applichiamo le funzioni count e mean
#alle colonne publish_time_dt, likes e dislike
interval_stats.head()
```

```python
#8. For each tag, determine the number of videos. Notice that tags contains a string with several tags.
#e
#9. Find the tags with the largest number of videos

# 1. Teniamo solo i video che hanno almeno un tag vero
df_tags = df2[df2["tags"] != "[none]"].copy()

# 2. Convertiamo la stringa in lista di tag
df_tags["tags_list"] = df_tags["tags"].apply(lambda x: x.split("|"))

# 3. Esplode: una riga per ogni tag
tags_exploded = df_tags.explode("tags_list")

# 4. Conta i tag
tag_counts = (
    tags_exploded
    .groupby("tags_list")
    .size()
    .sort_values(ascending=False)
)

tag_counts.head(20)
```

```python
#9. il tag con il maggior numero di video è 'funny'
```

```python
#10. For each (tag, country) pair, compute average ratio likes/dislikes
```

```python
tag_country_ratio = (
    tags_exploded
    .groupby(["tags_list", "country"])["like_ratio"]
    .mean()
    .reset_index()
    .sort_values(["tags_list", "country"])
)

tag_country_ratio.head(20)
```

```python
#cerchiamo ad esempio tutti i like_ratio del tag 'funny' per ogni paese

#utilizziamo un filtro con query
tag_country_ratio.query("tags_list == 'funny'")
```

```python
#12. Divide trending_date into three columns: year, month, day

#controlliamo prima il formato di tranding date e vediamo che non
#è un formato standard, ma YY.DD.MM (e non YYYY-MM-DD)

#dobbiamo dare prima delle indicazioni per leggere il formato correttamente

df2["trending_date"] = pd.to_datetime(df2["trending_date"], format="%y.%d.%m")

df2["year"] = df2["trending_date"].dt.year
df2["month"] = df2["trending_date"].dt.month
df2["day"] = df2["trending_date"].dt.day
```

```python
#15. For each country, determine how many videos have a category that is not assignable.

#in df2 abbiamo per ogni video category_id.

#nei json abbiamo dentro items, dentro snippet, 'assignable' :
#che puo essere True o False. 
```

```python
#costruiamo un df con le categorie per paese

df_cat = pd.DataFrame([
    {
        "country": country,
        "category_id": item["id"],
        "assignable": item["snippet"]["assignable"]
    }
    for country, content in json_data.items()
    for item in content["items"]
])

#convertiamo category_id in srtinga in entrambi i df per fare il merge
df2["category_id"] = df2["category_id"].astype(str)
df_cat["category_id"] = df_cat["category_id"].astype(str)

#uniamo con il merge i video e le categorie
df_merged = df2.merge(df_cat, on=["country", "category_id"], how="left")

#contiamo i video con le categorie non assegnabili:
not_assignable = (
    df_merged[df_merged["assignable"] == False]
    .groupby("country")
    .size()
    .reset_index(name="num_not_assignable")
)

not_assignable
```
